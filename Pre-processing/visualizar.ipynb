```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# carregando a base de dados
iris = datasets.load_iris()

# features
iris.feature_names
```

```python
# labels
iris.target
```

```python
iris.target_names
```

```python
# 
label = {0: 'red', 1: 'blue', 2: 'green'}

#dataset slicing

x_axis = iris.data[:, 0]  # Sepal Length
y_axis = iris.data[:, 2]  # Sepal Width

print("Sepal Length > ", type(x_axis), len(x_axis), x_axis.shape)
print(x_axis)
print("Sepal Width > ", type(y_axis), len(y_axis))
print(y_axis)
```

```python
# Plotting
plt.scatter(x_axis, y_axis, c=iris.target)
plt.title('Violet: Setosa, Green: Versicolor, Yellow: Virginica')
plt.show()
```

```python

def visualizar_clusters(kmeans, y_kmeans):
    
    plt.scatter(x_axis[y_kmeans == 0], y_axis[y_kmeans == 0],  c='red', label='Iris-setosa')
    plt.scatter(x_axis[y_kmeans == 1], y_axis[y_kmeans == 1],  c='blue', label='Iris-versicolour')
    plt.scatter(x_axis[y_kmeans == 2], y_axis[y_kmeans == 2],  c='yellow', label='Iris-virginica')

    # Plotting the centroids of the clusters
    plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100, c='green', label='Centroides')
    plt.legend()
    plt.show()

# declarando o modelo
model = KMeans(n_clusters=3)

# treinando o modelo para ajustar os centr√≥ides 
model.fit(iris.data)
 
# Prediction on the entire data
all_predictions = model.predict(iris.data)
print('All predictions = ', len(all_predictions ))

visualizar_clusters(model, all_predictions)
```
